```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.metrics.pairwise import cosine_distances
```

```python
def descrip_prox1(string, umbral = 0.4):
    
    # Crear la matriz de características TF-IDF para las descripciones existentes
    vectorizer = TfidfVectorizer()
    features = vectorizer.fit_transform(df_art['Nombre'])

    # Vectorizar la nueva descripción utilizando el mismo vectorizador TF-IDF
    nueva_caracteristica = vectorizer.transform([string])

    # Calcular la similitud de la nueva descripción con las existentes mediante la distancia del coseno
    # estas distancias serán valores entre 0 y 1
    distances = cosine_distances(nueva_caracteristica, features)

    # si la distancia es menor de la indicada en el umbral, se guarda en la lista "indices_cercanos"
    indices_cercanos = [i for i, distancia in enumerate(distances[0]) if distancia <= umbral]

    # Obtener las descripciones cercanas del DataFrame original
    descripciones_cercanas = df_art.iloc[indices_cercanos].Nombre.tolist()

    # print("Descripciones cercanas:")
    # for descripcion in descripciones_cercanas:
    #     print(descripcion)
    
    return descripciones_cercanas
```

```python
import Levenshtein

# def descrip_prox2(string, umbral=10):

#     # Calcular la distancia de Levenshtein entre la nueva descripción y cada descripción existente
#     distancias = [Levenshtein.distance(string, desc) for desc in df_art['Nombre']]

#     # Buscar índices de descripciones cercanas
#     indices_cercanos = [i for i, distancia in enumerate(distancias) if distancia <= umbral]

#     # Obtener las descripciones cercanas del DataFrame original
#     descripciones_cercanas = list(df_art.iloc[indices_cercanos]['Nombre'])

#     # Imprimir las descripciones cercanas
#     print("Descripciones cercanas:")
#     for descripcion in descripciones_cercanas:
#         print(descripcion)



def descrip_prox2(string, umbral_factor=0.1):
    # Calcular la distancia de Levenshtein entre la nueva descripción y cada descripción existente
    distancias = [Levenshtein.distance(string, desc) for desc in df_art['Nombre']]
    
    # Calcular la longitud promedio de las descripciones existentes
    longitud_promedio = len(df_art.Nombre.sum())/len(df_art)
    
    # Calcular el umbral relativo
    umbral = umbral_factor * longitud_promedio
    
    # Buscar índices de descripciones cercanas
    indices_cercanos = [i for i, distancia in enumerate(distancias) if distancia <= umbral]
    
    # Obtener las descripciones cercanas del DataFrame original
    descripciones_cercanas = list(df_art.iloc[indices_cercanos]['Nombre'])
    
    # print("Descripciones cercanas:")
    # for descripcion in descripciones_cercanas:
    #     print(descripcion)
    
    return descripciones_cercanas
```

```python
import jellyfish

def descrip_prox3(string, umbral=0.4):

    # Calcular la similitud de Jaro-Winkler entre la nueva descripción y las descripciones existentes
    descripciones = list(df_art['Nombre'])
    similitudes = [jellyfish.jaro_winkler(string, desc) for desc in descripciones]

    # Buscar índices de descripciones cercanas
    indices_cercanos = [i for i, similitud in enumerate(similitudes) if similitud >= umbral]

    # Obtener las descripciones cercanas del DataFrame original
    descripciones_cercanas = [descripciones[i] for i in indices_cercanos]

    # print("Descripciones cercanas:")
    # for descripcion in descripciones_cercanas:
    #     print(descripcion)
    
    return descripciones_cercanas
```

### la funcion "descrip_prox" toma un diccionario (nueva fila o registro que queremos añadir. las claves del diccionario serán las columnas del dataframe), el dataframe en el que vamos a buscar (debe contar al menos con las columnas 'Nombre' y 'Clase'), y los umbrales de las funciones de distancia que por defecto serán 0.9. 
Siempre devuelve un dataframe: con la nueva fila añadida y la clase del articulo como corresponda, o el dataframe sin modificacion alguna en caso de que finalmente so se añada el articulo. He querido hacerlo así para que se pueda sobreescribir siempre el dataframe con el output de la funcion (ya que si no lo hicieramos asi, en el caso de no añadir un articulo, si la funcion no devuelve nada y sobreescribimos el dataframe, tendremos problemas mas tarde).

Nota: el diccionario con el nuevo registro que le pasamos a la funcion, tendrá siempre el valor 0 en "Clase". Lo importante es asegurarse de no ponerle un valor mayor que el maximo de la columna "Clase".

```python
# primero voy a guardar df_art ya preprocesado en un csv
# df_art.to_csv("df_art.csv", index=False)
```

```python
def descrip_prox(dic, df, U1=0.9, U2=0.9):
    
    string = dic['Nombre']
    
    set1 = set(descrip_prox1(string, U1))
    set2 = set(descrip_prox2(string, U2))
    
    cercanos = list(set1 & set2)
    cercanos2 = list(set1 | set2)
    
    while True:
        if len(cercanos) == 0:
            if len(cercanos2) > 0:
                if len(cercanos2) > 4:
                    U1 -= 0.1
                    U2 -= 0.1
                    set1 = set(descrip_prox1(string, U1))
                    set2 = set(descrip_prox2(string, U2))
                    cercanos = list(set1 & set2)
                    cercanos2 = list(set1 | set2)
                    continue
                else:
                    inp = int(input(f'Existen algunos artículos con cierto parecido: "{cercanos2}". \nSi aún así desea añadirlo a la base de datos, introduzca: 1. \nEn caso contrario, introduzca: 0. \n'))
                    if inp == 0:
                        print('El nuevo artículo no se añadió finalmente.')
                        return df
                    elif inp == 1:
                        df = df.append(dic, ignore_index=True)
                        i = int(input('\nIntroduzca la posición en la lista del artículo que desea identificar con su nuevo artículo. \n'))
                        nombre = cercanos2[i-1]
                        a = df[df.Nombre == nombre].index
                        df.iloc[-1, -1] = df.iloc[a, -1]
                        return df
            else:
                df = df.append(dic, ignore_index=True)
                df.iloc[-1, -1] = df.Clase.max() + 1
                return df
        
        elif len(cercanos) == 1:
            cercano = str(*cercanos)
            inp = int(input(f'Ya existe un artículo muy parecido: "{cercano}". \nSi aún así desea añadirlo a la base de datos, introduzca: 1. \nEn caso contrario, introduzca: 0. \n'))
            if inp == 0:
                print('El nuevo artículo no se añadió finalmente.')
                return df
            elif inp == 1:
                df = df.append(dic, ignore_index=True)
                a = df[df.Nombre == cercano].index
                df.iloc[-1, -1] = df.iloc[a, -1]
                return df
        
        else:
            if len(cercanos) > 4:
                U1 -= 0.1
                U2 -= 0.1
                set1 = set(descrip_prox1(string, U1))
                set2 = set(descrip_prox2(string, U2))
                cercanos = list(set1 & set2)
                cercanos2 = list(set1 | set2)
                continue
            else:
                inp = int(input(f'Existen artículos muy similares: "{cercanos}". \nSi aún así desea añadirlo a la base de datos, introduzca: 1. \nEn caso contrario, introduzca: 0. \n'))
                if inp == 0:
                    print('El nuevo artículo no se añadió finalmente.')
                    return df
                elif inp == 1:
                    df = df.append(dic, ignore_index=True)
                    i = int(input('\nIntroduzca la posición en la lista del artículo que desea identificar con su nuevo artículo. \n'))
                    nombre = cercanos[i-1]
                    a = df[df.Nombre == nombre].index
                    df.iloc[-1, -1] = df.iloc[a, -1]
                    return df
```
